```python
from app.services.rag_service import RAGService

rag = RAGService()
rag.initialize()
print('RAG listo:', rag.is_ready(), '| chunks:', rag.chunk_count)
```

```python
queries = [
    'Cuales son los factores de riesgo cardiovascular?',
    'Como se prioriza un paciente con dolor toracico?',
    'Que hacer ante saturacion de oxigeno baja?',
]
for q in queries:
    docs = rag.query(q, n_results=3)
    print(f'\nQuery: {q}')
    if not docs:
        print('  (sin resultados)')
    for i, d in enumerate(docs, 1):
        content = d.get('content', '')
        print(f'  {i}. {content[:90]}...')
```
