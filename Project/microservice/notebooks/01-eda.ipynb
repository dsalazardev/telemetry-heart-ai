```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
from pathlib import Path
# Ejecutar siempre desde la raiz del microservicio.
if Path.cwd().name == 'notebooks':
    os.chdir('..')
print('cwd:', Path.cwd())
```

```python
df = pd.read_csv('app/data/synthetic_cases.csv')
print('Shape:', df.shape)
print('Columnas:', df.columns.tolist())
df.head()
```

```python
df.describe(include='all').T
```

```python
# Distribucion del target de triaje
counts = df['true_priority'].value_counts().sort_index()
labels = {0: 'BAJA', 1: 'MEDIA', 2: 'ALTA'}
plt.bar([labels.get(i, i) for i in counts.index], counts.values, color=['#2ca02c', '#ff7f0e', '#d62728'])
plt.title('Distribucion de true_priority')
plt.ylabel('n casos')
plt.show()
print(counts)
```

```python
# Matriz de correlacion (solo numericas)
num = df.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 10))
corr = num.corr()
plt.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)
plt.title('Matriz de correlacion')
plt.tight_layout()
plt.show()
```

```python
# Distribucion de las features clinicas crudas
clinical = ['heart_rate', 'spo2', 'systolic_bp', 'diastolic_bp', 'cholesterol', 'glucose', 'age']
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.ravel()
for idx, col in enumerate(clinical):
    axes[idx].hist(df[col], bins=20, edgecolor='black')
    axes[idx].set_title(col)
for j in range(len(clinical), len(axes)):
    axes[j].axis('off')
plt.tight_layout()
plt.show()
print('EDA completo')
```
